# meload dataset

```python
import pandas as pd
data = pd.read_csv("./dataset.csv")
```

# menampilkan dataset

```python
data
```

# analisis: jumlah pembeli pria dan wanita

```python
gender_counts = data['Gender'].value_counts()
print(gender_counts)
```

# analisis: barang yang paling suka dibeli oleh pria atau wanita

```python
most_purchased_items = data.groupby(['Gender', 'Item Purchased']).size().reset_index(name='Count')
most_purchased_items = most_purchased_items.sort_values(by='Count', ascending=False).groupby('Gender').head(1).reset_index(drop=True)
print(most_purchased_items)
```

# visualisasi: jumlah pembelian oleh pria dan wanita

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Count the number of males and females
gender_counts = data['Gender'].value_counts()

# Create a bar plot
sns.barplot(x=gender_counts.index, y=gender_counts.values)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Comparison of Men and Women')
plt.show()
```

# visualisasi: most bought items

```python
import seaborn as sns

most_bought_items = data['Item Purchased'].value_counts().reset_index()
most_bought_items.columns = ['Item Purchased', 'Count']

# Create a bar plot
sns.barplot(x='Item Purchased', y='Count', data=most_bought_items)
plt.xlabel('Item Purchased')
plt.ylabel('Count')
plt.title('Most Bought Items')
plt.xticks(rotation=90)
plt.show()
```

# analisis: korelasi pearson antara musim dan rating

```python
correlation = data['Purchase Amount (USD)'].corr(data['Review Rating'])
print(f"Pearson correlation between Purchase Amount (USD) and Review Rating: {correlation}")
```

# Transormasi data kategorical menjadi data numeral

```python
# Define a mapping for sizes
size_mapping = {'XS': 1, 'S': 2, 'M': 3, 'L': 4, 'XL': 5}

# Apply the mapping to the Size column
data['Size'] = data['Size'].map(size_mapping)

# Display the updated dataframe
data.head()
```

# Analisis: Apakah Ukuran benda Clothing yang dibeli, berkolerasi dengan rating

```python
# Filter the data for the "Clothing" category
clothing_data = data[data['Category'] == 'Clothing'] # mengandung hanya data baju

# Calculate the correlation between review rating and size
clothing_correlation = clothing_data['Review Rating'].corr(clothing_data['Size'])
print(f"Correlation between Review Rating and Size for Clothing: {clothing_correlation}")
```
